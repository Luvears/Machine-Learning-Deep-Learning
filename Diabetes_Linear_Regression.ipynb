### 1. Data Loading and Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df_diabetes['target'] = diabetes.target

print('Shape:' , df_diabetes.shape)
print('\n' + '=='*25 + "\n")
print('\nColumn Names:', df_diabetes.columns)
print('\n' + '=='*25 + "\n")
print(df_diabetes.head())
print('\n' + '=='*25 + "\n")
print(df_diabetes.describe())
print('\n' + '=='*25 + "\n")
print(diabetes.DESCR[:1500])
```

### 2. Data Visualization

```python
import seaborn as sns

fig1, axes1 = plt.subplots(5, 2, figsize=(14, 30))
fig1.suptitle('Diabetes Data: Feature Relationship with Target Value' , fontsize=16)


#Plot 1: age
axes1[0,0].scatter(df_diabetes['age'], df_diabetes['target'] , alpha=0.4 ,s=10)
axes1[0,0].set_xlabel('Age')
axes1[0,0].set_ylabel('Target')
axes1[0,0].set_title('Age vs Target')
axes1[0,0].grid(True, alpha = 0.3)

#Plot 2: sex
sns.violinplot(x=df_diabetes['sex'], y=df_diabetes['target'] , ax=axes1[0,1], color='green' )
axes1[0,1].set_xlabel('Sex')
axes1[0,1].set_ylabel('Target')
axes1[0,1].set_title('Sex vs Target')
axes1[0,1].grid(True, alpha = 0.3)

#Plot 3: bmi
axes1[1,0].scatter(df_diabetes['bmi'], df_diabetes['target'] , alpha=0.4 ,s=10, color='red')
axes1[1,0].set_xlabel('bmi')
axes1[1,0].set_ylabel('Target')
axes1[1,0].set_title('bmi vs Target')
axes1[1,0].grid(True, alpha = 0.3)

#Plot 4: bp
axes1[1,1].scatter(df_diabetes['bp'], df_diabetes['target'] , alpha=0.4 ,s=10, color='purple')
axes1[1,1].set_xlabel('bp')
axes1[1,1].set_ylabel('Target')
axes1[1,1].set_title('bp vs Target')
axes1[1,1].grid(True, alpha = 0.3)

#Plot 5: s1
axes1[2,0].scatter(df_diabetes['s1'], df_diabetes['target'] , alpha=0.4 ,s=10, color='black')
axes1[2,0].set_xlabel('tc (total serum cholesterol)')
axes1[2,0].set_ylabel('Target')
axes1[2,0].set_title('tc vs Target')
axes1[2,0].grid(True, alpha = 0.3)

#Plot 6: s2
axes1[2,1].scatter(df_diabetes['s2'], df_diabetes['target'] , alpha=0.4 ,s=10, color='gold')
axes1[2,1].set_xlabel('ldl (low-density lipoproteins)')
axes1[2,1].set_ylabel('Target')
axes1[2,1].set_title('ldl vs Target')
axes1[2,1].grid(True, alpha = 0.3)

#Plot 7: s3
axes1[3,0].scatter(df_diabetes['s3'], df_diabetes['target'] , alpha=0.4 ,s=10, color='pink')
axes1[3,0].set_xlabel('hdl (high-density lipoproteins)')
axes1[3,0].set_ylabel('Target')
axes1[3,0].set_title('hdl vs Target')
axes1[3,0].grid(True, alpha = 0.3)

#Plot 8: s4
axes1[3,1].scatter(df_diabetes['s4'], df_diabetes['target'] , alpha=0.4 ,s=10, color='brown')
axes1[3,1].set_xlabel('tch (total cholesterol)')
axes1[3,1].set_ylabel('Target')
axes1[3,1].set_title('tch vs Target')
axes1[3,1].grid(True, alpha = 0.3)

#Plot 9: s5
axes1[4,0].scatter(df_diabetes['s5'], df_diabetes['target'] , alpha=0.4 ,s=10, color='blue')
axes1[4,0].set_xlabel('ltg (possibly log of serum triglycerides level)')
axes1[4,0].set_ylabel('Target')
axes1[4,0].set_title('ltg vs Target')
axes1[4,0].grid(True, alpha = 0.3)

#Plot 10: s6
axes1[4,1].scatter(df_diabetes['s6'], df_diabetes['target'] , alpha=0.4 ,s=10, color='green')
axes1[4,1].set_xlabel('glu (blood sugar level)')
axes1[4,1].set_ylabel('Target')
axes1[4,1].set_title('glu vs Target')
axes1[4,1].grid(True, alpha = 0.3)



plt.figure(figsize=(6,4))
plt.hist(df_diabetes['target'], bins=40, edgecolor='black', color='purple')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.title('Distribution of The Target')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("\n Observation: Notice positive correlation between bmi/ltg and the target.")
```

### 3. Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_diabetes = diabetes.data
Y_diabetes = diabetes.target

X_train, X_test, Y_train, Y_test = train_test_split(
    X_diabetes, Y_diabetes,
    test_size=0.2,
    random_state=42)

print(f"Training set - Features: {X_train.shape}, Target: {Y_train.shape}")
print(f"Test set - Features: {X_test.shape}, Target: {Y_test.shape}")

```

### 4. Model Training

```python
# Model 1: Single best feature

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


print("Training with single feature s5: ltg(log of serum triglycerides level)")

print('\n' + '=='*25 + "\n")

X_simple = X_train[:,8].reshape(-1,1)
X_simple_test = X_test[:,8].reshape(-1,1)
print(f'Training Set shape: {X_simple.shape}')
print(f'Test Set Shape: {X_simple_test.shape}')

print('\n' + '=='*25 + "\n")

model_simple = LinearRegression()
model_simple.fit(X_simple, Y_train)
print(" Model trained")

print('\n' + '=='*25 + "\n")

print("Learned Parameters:")
print(f"Intercept (): {model_simple.intercept_:.4f}")
print(f"Slope (): {model_simple.coef_[0]:.4f}")
print(f"\nEquation: Target = {model_simple.intercept_:.4f} + {model_simple.coef_[0]:.4f}  s5")
```

```python
# Model 2: All features

print("Multiple Linear Regression Model")
print('\n' + '=='*25 + "\n")

model_multiple = LinearRegression()
model_multiple.fit(X_train, Y_train)

print(" Model trained\n")

print(f"\nIntercept (): {model_multiple.intercept_:.4f}")
print("\nFeature Coefficients:")
for feature_name, coef in zip(diabetes.feature_names, model_multiple.coef_):
    print(f"  {feature_name:12s}: {coef:8.4f}")
```

```python
# Mode 3: Selected features: bmi, bp, hdl, ltg

print("Training with selected feature: bmi, bp, hdl, ltg)")
print('\n' + '=='*25 + "\n")

features = ['bmi','bp','s3','s5']

X_slct = df_diabetes[['bmi', 'bp', 's3', 's5']]
Y_slct = df_diabetes['target']

X_slct_train, X_slct_test, Y_slct_train, Y_slct_test = train_test_split(
    X_slct, Y_slct,
    test_size=0.2,
    random_state=42)


model_slct = LinearRegression()
model_slct.fit(X_slct_train, Y_slct_train)

print(" Model trained\n")

print(f"\nIntercept (): {model_slct.intercept_:.4f}")
print("\nFeature Coefficients:")

for feature_name, coef in zip(features, model_slct.coef_):
    print(f"  {feature_name:12s}: {coef:8.4f}")


```

### 5. Model Evaluation

```python
Y_pred_train = model_simple.predict(X_simple)
Y_pred_test = model_simple.predict(X_simple_test)

train_rmse = np.sqrt(mean_squared_error(Y_train, Y_pred_train))
test_rmse = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
train_r2 = r2_score(Y_train, Y_pred_train)
test_r2 = r2_score(Y_test, Y_pred_test)

print("Model Performance (s5 Feature):")
print(f"\nTraining Set:")
print(f"  RMSE: {train_rmse:.4f}")
print(f"  R Score: {train_r2:.4f}")

print(f"\nTest Set:")
print(f"  RMSE: {test_rmse:.4f} ")
print(f"  R Score: {test_r2:.4f}")


print('\n' + '=='*25 + "\n")


Y_pred_train_multi = model_multiple.predict(X_train)
Y_pred_test_multi = model_multiple.predict(X_test)

train_rmse_multi = np.sqrt(mean_squared_error(Y_train, Y_pred_train_multi))
test_rmse_multi = np.sqrt(mean_squared_error(Y_test, Y_pred_test_multi))
train_r2_multi = r2_score(Y_train, Y_pred_train_multi)
test_r2_multi = r2_score(Y_test, Y_pred_test_multi)

print("Model Performance (All Features):")
print(f"\nTraining Set:")
print(f"  RMSE: {train_rmse_multi:.4f}")
print(f"  R Score: {train_r2_multi:.4f}")

print(f"\nTest Set:")
print(f"  RMSE: {test_rmse_multi:.4f}")
print(f"  R Score: {test_r2_multi:.4f}")



print('\n' + '=='*25 + "\n")


Y_pred_train_slct = model_slct.predict(X_slct_train)
Y_pred_test_slct = model_slct.predict(X_slct_test)

train_rmse_slct = np.sqrt(mean_squared_error(Y_slct_train, Y_pred_train_slct))
test_rmse_slct = np.sqrt(mean_squared_error(Y_slct_test, Y_pred_test_slct))
train_r2_slct = r2_score(Y_slct_train, Y_pred_train_slct)
test_r2_slct = r2_score(Y_slct_test, Y_pred_test_slct)

print("Model Performance (Selected Features):")
print(f"\nTraining Set:")
print(f"  RMSE: {train_rmse_slct:.4f}")
print(f"  R Score: {train_r2_slct:.4f}")

print(f"\nTest Set:")
print(f"  RMSE: {test_rmse_slct:.4f}")
print(f"  R Score: {test_r2_slct:.4f}")
```

### 6. Visualization of Results

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.scatter(Y_test, Y_pred_test, alpha=0.3, s=20)
plt.plot([Y_test.min() , Y_test.max()] , [Y_pred_test.min() , Y_pred_test.max()] , 'r--',linewidth=2 , label='Perferct Prediction')
plt.xlabel('actual target')
plt.ylabel('predicted target')
plt.title(f'one feature model (s5)\n(R = {test_r2:.4f})')
plt.legend()
plt.grid(True, alpha=0.3)


plt.subplot(1, 3, 2)
plt.scatter(Y_test, Y_pred_test_multi, alpha=0.3, s=20)
plt.plot([Y_test.min() , Y_test.max()] , [Y_pred_test_multi.min() , Y_pred_test_multi.max()] , 'r--',linewidth=2 , label='Perferct Prediction')
plt.xlabel('actual target')
plt.ylabel('predicted target')
plt.title(f'All feature model\n(R = {test_r2_multi:.4f})')
plt.legend()
plt.grid(True, alpha=0.3)


plt.subplot(1, 3, 3)
plt.scatter(Y_test, Y_pred_test_multi, alpha=0.3, s=20)
plt.plot([Y_test.min() , Y_test.max()] , [Y_pred_test_slct.min() , Y_pred_test_slct.max()] , 'r--',linewidth=2 , label='Perferct Prediction')
plt.xlabel('actual target')
plt.ylabel('predicted target')
plt.title(f'All feature model\n(R = {test_r2_slct:.4f})')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
